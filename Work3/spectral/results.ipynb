```python
# Import Libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Results
results_path = '../results/spectral_results.csv'
df = pd.read_csv(results_path)
df.head()
```

### **Data Preprocessing**

```python
# Convert numerical columns to float
numeric_cols = ['Silhouette', 'ARI', 'DBI', 'Time (s)']
df[numeric_cols] = df[numeric_cols].astype(float)
```

### **Plotting**

#### **ARI vs N Neighbors**

```python
# List of datasets
datasets = df['Dataset'].unique()

# Set the plotting style
sns.set(style='whitegrid')

# Create a color palette
palette = sns.color_palette("husl", len(df['Eigen Solver'].unique()))

# Initialize the matplotlib figure
fig, axes = plt.subplots(1, len(datasets), figsize=(18, 6), sharey=True)

for ax, dataset in zip(axes, datasets):
    subset = df[df['Dataset'] == dataset]
    sns.lineplot(
        data=subset,
        x='N Neighbors',
        y='ARI',
        hue='Eigen Solver',
        marker='o',
        palette=palette,
        ax=ax
    )
    ax.set_title(f'{dataset.capitalize()} Dataset')
    ax.set_xlabel('N Neighbors')
    ax.set_ylabel('Adjusted Rand Index (ARI)')
    ax.legend(title='Eigen Solver')
    
plt.tight_layout()
plt.show()
```
